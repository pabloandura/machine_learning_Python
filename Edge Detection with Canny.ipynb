# Edge Detection 

```python
# We will use the Canny edge detector
import cv2
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
img = cv2.imread('Computer-Vision-with-Python/DATA/sammy_face.jpg')
```

```python
plt.imshow(img)
```

```python
edge = cv2.Canny(image=img,threshold1=127,threshold2=127)
plt.imshow(edge)
```

```python
edge = cv2.Canny(image=img,threshold1=0,threshold2=255)
plt.imshow(edge)
```

```python
edge = cv2.Canny(image=img,threshold1=255,threshold2=255)
plt.imshow(edge)
```

```python
# lets calculate the median pixel value
med_val = np.median(img)
med_val
```

```python
# lower threshold to either 0 or 70% of the median value whichever is greater
lower = int(max(0,0.7*med_val))
# upper treshhold to either 130% or the max 255 whichever is smaller
upper = int(min(255,1.3*med_val))
```

```python
edge = cv2.Canny(image=img,threshold1=lower,threshold2=upper+100)
plt.imshow(edge)
# it did worse so we increased the upper by 100

# we will blur the image 
blurred = cv2.blur(img,ksize=(5,5))
edge = cv2.Canny(image=blurred,threshold1=lower,threshold2=upper)
plt.imshow(edge)
```
