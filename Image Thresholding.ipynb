```python
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img = cv2.imread('Computer-Vision-with-Python/DATA/rainbow.jpg',0)
plt.imshow(img,cmap='gray')
```

```python
# turn this picture into binary
ret,thresh1 = cv2.threshold(img,127,255,cv2.THRESH_BINARY)
# returns cutoff value and an image
ret
```

```python
plt.imshow(thresh1,cmap='gray')
```

```python
# as we can see the image has two different "colors"
img = cv2.imread('Computer-Vision-with-Python/DATA/crossword.jpg',0)
```

```python
plt.imshow(img,cmap='gray')
```

```python
# set up a function to display image
def show_pic(img):
    fig=plt.figure(figsize=(15,15))
    ax = fig.add_subplot(111)
    ax.imshow(img,cmap='gray')

show_pic(img)
```

```python
# Lets set up a binary threshold to make our crossword easier to read
ret, th1 = cv2.threshold(img,127,255,cv2.THRESH_BINARY)
show_pic(th1)
```

```python
# We lost information, lets experiment with a different threshold level

ret, th1 = cv2.threshold(img,150,255,cv2.THRESH_BINARY)
show_pic(th1)
```

```python
# another method is the adaptive threshold
# there is a full notebook to explore this technique

th2 = cv2.adaptiveThreshold(img,255,cv2.ADAPTIVE_THRESH_MEAN_C,cv2.THRESH_BINARY,11,8)
show_pic(th2)
# We have 2 more parameters, a block size and a constant value
```

```python
# Lets change the values slightly
th2 = cv2.adaptiveThreshold(img,255,cv2.ADAPTIVE_THRESH_MEAN_C,cv2.THRESH_BINARY,11,14)
show_pic(th2)
```
